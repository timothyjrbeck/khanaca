```python
import requests
from bs4 import BeautifulSoup
import urllib.request
from pathlib import Path
import json

base_url = "https://www.khanacademy.org"

def html2script(soup):
    script = str(soup.find_all('script')[4])
    start = script.find('window["./javascript/app-shell-package/app-entry.js"] = ') + len('window["./javascript/app-shell-package/app-entry.js"] = ')
    end = script.find(';\n    </script>')
    script = script.replace('true', "True")
    script = script.replace('false', "False")
    script = script.replace('null', "None")
    return eval(script[start:end])

# Subject level
# for module in units:
#     print(module['url'])
def get_modules(subject):
    r = requests.get(base_url + subject)
    soup = BeautifulSoup(r.text, 'html.parser')
    script = html2script(soup)
    return script['dataCache']['content-en'][list(script['dataCache']['content-en'].keys())[0]]['data']['tabs'][0]['modules'][1]['modules']


# Unit level
# soup.find_all(attrs={"data-test-id":"unit-header"})
def get_units(unit):
    r = requests.get(base_url + unit)
    soup = BeautifulSoup(r.text, 'html.parser')
    script = html2script(soup)
    modules = script['dataCache']['content-en'][list(script['dataCache']['content-en'].keys())[0]]['data']['tabs'][0]['modules'][0]
    return modules['tutorials']

# Video download
def video_dl(item,folder_dir):
    urllib.request.urlretrieve(item['downloadUrls']['mp4'], f"./dl{folder_dir}/{item['slug']}.mp4") 
# "https://cdn.kastatic.org/ka-youtube-converted/Vi1JK6IYVt8.mp4/Vi1JK6IYVt8.mp4"

# Page level: Article

# for i in list(content.keys()):
#     print(i)
#     print(content[i])
#     print("\n\n\n")

def article_dl(item,folder_dir):
    r = requests.get(base_url+item['nodeUrl'])
    soup = BeautifulSoup(r.text, 'html.parser')
    script = html2script(soup)
    content = script['dataCache']['content-en'][list(script['dataCache']['content-en'].keys())[0]]['data']['tutorialPageData']['contentModel']
    data = eval(content['perseusContent'])
    with open(f"./dl{folder_dir}/{item['slug']}.json", 'w') as fp:
        json.dump(data, fp)

def content_dl(i,folder_dir):
    if(i['kind'] == 'Video'):
        print("Video")
        video_dl(i,folder_dir)
    if(i['kind'] == 'Article'):
        print("Article")
        article_dl(i,folder_dir)

def unit_dl(unit):
    folder_dir = unit['url'].replace('#',"/")
    Path("./dl/"+folder_dir).mkdir(parents=True, exist_ok=True)
    print("NEW UNIT")
    print(unit['title'])
    for i in unit['contentItems']:
        content_dl(i,folder_dir)
    print('')

def module_dl(module):
    print("NEW MODULE")
    print(module['title'])
    units = get_units(module['url'])
    for unit in units:
        unit_dl(unit)
    print("\n")
```
